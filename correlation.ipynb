let first import pandas library and the school dataset

````python
import pandas as pd
df=pd.read_csv('data/school.csv')
df.head()
````

Now lets calculate all the correlation in the dataset 

````python
df.corr()
````

You can see that the ```corr()```calculate the correlation  matrix of all the numerical columns in the dataframe
We can notice that their are :

* a small positive correlation :

  * age-weight(its normal when your age increase,your weight increase)
  * age-height (its normal when your age increase,your height growht) 
* a higher positive correlation :
  * weight-height (we have a higher correlation than age-weight or age-height because the physical growth is a more important factor of weight gain than age.
